```python
import requests
import os
import json
import pandas as pd
import math

def clean_text(text):
    if text is None:
        return None

    # Kill NaN floats
    if isinstance(text, float) and math.isnan(text):
        return None

    # Force string
    text = str(text)

    # Kill string NaN
    if text.strip().lower() == "nan":
        return None

    # Remove weird whitespace
    text = text.strip()

    if len(text) == 0:
        return None

    return text

def create_embedding(texts, batch_size=16):
    all_embeddings = []

    for i in range(0, len(texts), batch_size):
        batch = texts[i:i + batch_size]

        try:
            r = requests.post(
                "http://localhost:11434/api/embed",
                json={
                    "model": "bge-m3",
                    "input": batch
                },
                timeout=120
            )

            response = r.json()

            if "error" in response:
                raise RuntimeError(response["error"])

            if "embeddings" in response:
                all_embeddings.extend(response["embeddings"])
            elif isinstance(response, list):
                all_embeddings.extend([item["embedding"] for item in response])
            else:
                raise RuntimeError(f"Unexpected response: {response}")

        except Exception as e:
            print(f"âš ï¸ Batch failed at index {i}, retrying individually...")

            # ðŸ”¥ FALLBACK: embed one-by-one
            for j, text in enumerate(batch):
                try:
                    r = requests.post(
                        "http://localhost:11434/api/embed",
                        json={
                            "model": "bge-m3",
                            "input": [text]
                        },
                        timeout=120
                    )

                    response = r.json()

                    if "embeddings" in response:
                        all_embeddings.append(response["embeddings"][0])
                    elif isinstance(response, list):
                        all_embeddings.append(response[0]["embedding"])
                    else:
                        raise RuntimeError(response)

                except Exception:
                    print(f"âŒ Skipping bad chunk at global index {i + j}")
                    # push a zero-vector placeholder to keep alignment
                    all_embeddings.append([0.0] * 1024)

    return all_embeddings


# ------------------ MAIN ------------------

json_dir = "jsons"
all_chunks = []
chunk_id = 0

for json_file in os.listdir(json_dir):
    with open(os.path.join(json_dir, json_file), "r", encoding="utf-8") as f:
        content = json.load(f)

    print(f"Creating Embeddings for {json_file}")

    texts = [clean_text(c.get("text")) for c in content["chunks"]]
    texts = [t for t in texts if t]

    if not texts:
        print("âš ï¸ Skipping file â€” no valid text")
        continue


    embeddings = create_embedding(texts)

    if len(embeddings) != len(texts):
        raise RuntimeError("âŒ Embedding count mismatch")

    for i, chunk in enumerate(content["chunks"]):
        if not clean_text(chunk.get("text")):
            continue

        chunk["chunk_id"] = chunk_id
        chunk["embedding"] = embeddings.pop(0)
        chunk_id += 1
        all_chunks.append(chunk)
        
for i, c in enumerate(content["chunks"]):
    if clean_text(c.get("text")) is None:
        print("BAD CHUNK AT:", i, repr(c.get("text")))

df = pd.DataFrame(all_chunks)
print(df.head())

```
